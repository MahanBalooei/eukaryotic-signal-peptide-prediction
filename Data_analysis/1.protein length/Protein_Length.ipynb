```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def load_and_preprocess(tsv_file):
    """Load TSV, convert types, and add 'Class' for plotting."""
    df = pd.read_csv(tsv_file, sep='\t')
    df.columns = df.columns.str.strip()  # Clean names
    df['SP cleavage'] = pd.to_numeric(df['SP cleavage'], errors='coerce')
    df['label'] = pd.to_numeric(df['label'], errors='coerce').fillna(0).astype(int)
    df['Class'] = df['label'].map({1: 'Positive (with SP)', 0: 'Negative (no SP)'})
    print(f"Loaded {len(df)} entries from {tsv_file}")
    return df

def plot_protein_lengths(df, dataset_name):
    """Density plot with shortened x-axis to highlight overlap and remove empty tails."""
    plt.figure(figsize=(10, 6))
    sns.kdeplot(data=df[df['label'] == 1], x='Sequence length', fill=True, label='Positive (with SP)', color='blue')
    sns.kdeplot(data=df[df['label'] == 0], x='Sequence length', fill=True, label='Negative (no SP)', color='red')
    plt.title(f'Protein Length Distribution - {dataset_name} (Positive vs Negative)')
    plt.xlabel('Protein Length')
    plt.ylabel('Density')
    plt.xlim(0, 2000)  # Shortened domain for visibility of overlap; adjust if needed (e.g., based on df['Sequence length'].quantile(0.99))
    plt.legend()
    plt.savefig(f'{dataset_name.lower()}_protein_lengths.png')
    plt.close()
    print(f"Saved plot: {dataset_name.lower()}_protein_lengths.png")

# Run for training
train_df = load_and_preprocess('training_with_folds.tsv')
plot_protein_lengths(train_df, 'Training')

# Run for test (benchmarking)
test_df = load_and_preprocess('test_set_labeled.tsv')
plot_protein_lengths(test_df, 'Test')
```
