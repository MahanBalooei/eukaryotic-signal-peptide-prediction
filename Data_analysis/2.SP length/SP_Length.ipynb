```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def load_positives(tsv_file):
    """Load TSV and filter positives (with SP)."""
    df = pd.read_csv(tsv_file, sep='\t')
    df.columns = df.columns.str.strip()
    df['SP cleavage'] = pd.to_numeric(df['SP cleavage'], errors='coerce')
    df['label'] = pd.to_numeric(df['label'], errors='coerce').fillna(0).astype(int)
    positives = df[df['label'] == 1].copy()
    positives['SP_length'] = positives['SP cleavage'] - 1  # Compute SP length
    print(f"Loaded {len(positives)} positives from {tsv_file}")
    return positives

def plot_sp_lengths(pos_df, dataset_name):
    """Histogram with KDE for SP length distribution."""
    plt.figure(figsize=(10, 6))
    sns.histplot(pos_df['SP_length'].dropna(), bins=20, kde=True, color='green')
    plt.title(f'SP Length Distribution - {dataset_name}')
    plt.xlabel('SP Length')
    plt.ylabel('Frequency')
    plt.savefig(f'{dataset_name.lower()}_sp_lengths.png')
    plt.close()
    print(f"Saved plot: {dataset_name.lower()}_sp_lengths.png")

# Run for training
train_pos = load_positives('training_with_folds.tsv')
plot_sp_lengths(train_pos, 'Training')

# Run for test (benchmarking)
test_pos = load_positives('test_set_labeled.tsv')
plot_sp_lengths(test_pos, 'Test')
```
