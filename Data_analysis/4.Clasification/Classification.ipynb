```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

def load_data(tsv_file):
    """Load TSV for taxonomy."""
    df = pd.read_csv(tsv_file, sep='\t')
    df.columns = df.columns.str.strip()
    print(f"Loaded {len(df)} entries from {tsv_file}")
    return df

def plot_taxonomy(df, col, dataset_name, top_n=None):
    """Barplot for kingdom or species."""
    counts = df[col].value_counts()
    if top_n:
        counts = counts.head(top_n)
        title = f'Top {top_n} Species - {dataset_name}'
    else:
        title = f'{col} Distribution - {dataset_name}'
    plt.figure(figsize=(10, max(6, len(counts) * 0.3)))
    sns.barplot(x=counts.values, y=counts.index, palette='viridis')
    plt.title(title)
    plt.xlabel('Count')
    plt.savefig(f'{dataset_name.lower()}_{col.lower()}.png')
    plt.close()
    print(f"Saved plot: {dataset_name.lower()}_{col.lower()}.png")

# Run for training
train_df = load_data('training_with_folds.tsv')
plot_taxonomy(train_df, 'Kingdom', 'Training')
plot_taxonomy(train_df, 'Organism', 'Training', top_n=10)

# Run for test
test_df = load_data('test_set_labeled.tsv')
plot_taxonomy(test_df, 'Kingdom', 'Test')
plot_taxonomy(test_df, 'Organism', 'Test', top_n=10)
```
